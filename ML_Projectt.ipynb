```python
from sklearn.model_selection import train_test_split
```

```python
def converter(ref):
    if ref=="yes":
        return 1
    else:
        return 0
    
```

```python
df["voice plan"]=df["voice mail plan"].apply(converter)
df["intl plan"]=df["international plan"].apply(converter)
df.drop("voice mail plan",axis=1,inplace=True)
df.drop("international plan",axis=1,inplace=True)
df.head(15)
```

```python
X=df[["area code","total day minutes","total day charge","total eve minutes","total eve charge","total night minutes",
      "total night charge","total intl minutes","total intl charge","customer service calls",
      "voice plan","total charges"]]
#X=df[["customer service calls"]]
y=df["churn"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
print(predictions)
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```
